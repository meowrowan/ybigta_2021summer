```python
import re
import requests
from bs4 import BeautifulSoup
```

```python
cine21_url = 'http://www.cine21.com/rank/person/content'
post_data = dict()
post_data['section'] = 'actor'
post_data['period_start'] = '2021-07'
post_data['gender'] = 'all'

actors_info_list = list()

for i in range(1, 11):
  post_data['page'] = i
  res = requests.post(cine21_url, data=post_data)
  soup = BeautifulSoup(res.text, 'html.parser')

  actors = soup.select('li.people_li div.name')
  nums = soup.select('li.people_li > ul.num_info > li')
  mov_lists = soup.select('ul.mov_list span')
  ranks = soup.select('span.grade')

  for index, actor in enumerate(actors):
    actor_info_dict = dict()

    actor = actors[index].text
    actor_name = re.sub('\(\w*\)', '', actor)
    actor_info_dict['이름'] = actor_name
    num = re.sub('[^0-9]', '', nums[index].text)
    actor_info_dict['흥행지수'] = num
    movies =[]
    for mov in mov_lists[index].text.split("\n"):
      if mov.strip():
        movies.append(mov)
    actor_info_dict['영화목록'] = movies
    rank = ranks[index].text
    actor_info_dict['랭킹'] = rank
    
    actor_link = 'http://www.cine21.com' + actors[index].a['href']
    response_actor = requests.get(actor_link)
    soup_actor = BeautifulSoup(response_actor.text, 'html.parser')
    actor_details = soup_actor.select_one('ul.default_info').select('li')
    
    for actor_detail in actor_details:
        actor_detail_key = actor_detail.select_one('span.tit').text
        actor_detail_value = re.sub('<span.*>.*</span>', '', str(actor_detail))
        actor_detail_value = re.sub('<.*?>', '', actor_detail_value)
        actor_info_dict[actor_detail_key] = actor_detail_value
        
    actors_info_list.append(actor_info_dict)

actors_info_list

```
